## <b>Image Pyramid <b>

### 1. Pyramid Down with cv2.pyrDown()

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (use correct local path)
image = cv2.imread('p1.webp')

# Check if image loaded properly
if image is None:
    print("Image not found. Check the file path.")
else:
    # Downsample image (reduce size by half)
    downsampled_image = cv2.pyrDown(image)

    # Convert BGR  RGB for correct color display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    downsampled_rgb = cv2.cvtColor(downsampled_image, cv2.COLOR_BGR2RGB)

    # Display images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image_rgb)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Downsampled Image (pyrDown)")
    plt.imshow(downsampled_rgb)
    plt.axis("off")

    plt.show()
```

### 2. Pyramid Up with cv2.pyrUp()

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (use correct local path)
image = cv2.imread('p1.webp')

# Check if image loaded
if image is None:
    print("Image not found. Check the file path.")
else:
    # Upsample image (double size)
    upsampled_image = cv2.pyrUp(image)

    # Convert BGR  RGB for correct display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    upsampled_rgb = cv2.cvtColor(upsampled_image, cv2.COLOR_BGR2RGB)

    # Display images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image_rgb)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Upsampled Image (pyrUp)")
    plt.imshow(upsampled_rgb)
    plt.axis("off")

    plt.show()
```

### Building a Gaussian Pyramid (Multiple Levels)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (use correct local path)
image = cv2.imread('p1.webp')

# Check if image loaded
if image is None:
    print("Image not found. Check the file path.")
else:
    # Create pyramid list
    pyramid = [image]

    # Build pyramid (3 levels down)
    for i in range(3):
        image = cv2.pyrDown(image)
        pyramid.append(image)

    # Display pyramid from smallest to largest
    plt.figure(figsize=(12, 8))

    for i in range(len(pyramid) - 1, -1, -1):
        img_rgb = cv2.cvtColor(pyramid[i], cv2.COLOR_BGR2RGB)

        plt.subplot(1, len(pyramid), len(pyramid) - i)
        plt.title(f"Pyramid Level {i}")
        plt.imshow(img_rgb)
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```
